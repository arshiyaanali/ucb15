```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('coupons.csv')
df
```

```python
df.describe() #gives us the statistical data analysis of the dataframe columns, that is, the mean, median, and std
```

```python
df.info() #gives us the datatypes of the columns in the dataframe
```

```python
#Gives us a bar plot of both Accepted and Not Accepted Coupon by Venue
plt.figure(figsize=(10, 6))
sns.countplot(x='coupon', hue='Y', data=df)
plt.title('Coupon Acceptance by Coupon Type')
plt.show()
```

```python
# Display summaries and compare whether coupons are accepted vs when they are not excepted
accepted_summary = df[df['Y'] == 1].describe()
not_accepted_summary = df[df['Y'] == 0].describe()

print("Accepted Coupons Summary:")
print(accepted_summary)
print("Not Accepted Coupons Summary:")
print(not_accepted_summary)
```

```python
# Distribution of Coupon Types at Each Venue as a Pie Chart so we can see percentages
plt.figure(figsize=(8, 8))
coupon_counts = df['coupon'].value_counts()
plt.pie(coupon_counts, labels=coupon_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Coupon Types')
plt.show()
```

```python
#These lines of code tell us the Probability of the Coupon being accepted as a Pie Chart
plt.figure(figsize=(16, 6))
plt.subplot(1, 3, 1)
coupon_acceptance_counts = df['Y'].value_counts()
labels_acceptance = ['Not Accepted', 'Accepted']
plt.pie(coupon_acceptance_counts, labels=labels_acceptance, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Coupon Acceptance')

#These lines of code tells us the Probability of the Coupon being accepted at each venue as a Pie Chart
plt.subplot(1, 3, 2)
accepted_coupon_types_counts = df[df['Y'] == 1]['coupon'].value_counts()
labels_accepted_types = accepted_coupon_types_counts.index
plt.pie(accepted_coupon_types_counts, labels=labels_accepted_types, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Coupon Types (Among Accepted Coupons)')

#These lines of code tells us the Probability of the Coupon not being accepted at each venue as a Pie Chart
plt.subplot(1, 3, 3)
not_accepted_coupon_types_counts = df[df['Y'] == 0]['coupon'].value_counts()
labels_not_accepted_types = not_accepted_coupon_types_counts.index
plt.pie(not_accepted_coupon_types_counts, labels=labels_not_accepted_types, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Coupon Types (Among Not Accepted Coupons)')

plt.tight_layout()
plt.show()
```
